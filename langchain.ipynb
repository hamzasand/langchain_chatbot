### ChatBot with chatprompt template


```python
import os
from dotenv import load_dotenv
load_dotenv()
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
GOOGLE_API_KEY=os.getenv("GOOGLE_API_KEY")
LANGCHAIN_TRACING_V2=os.getenv("LANGCHAIN_TRACING_V2")
LANGCHAIN_API_KEY=os.getenv("LANGCHAIN_API_KEY")
```

#### Set model

```python
model = ChatGoogleGenerativeAI(model="gemini-3.1-flash-lite", temperature= 0.2)
```

#### Set chatprompttemplate

```python
prompt_template = ChatPromptTemplate(
    [
        ("system","Translate foillowing into {language}"),
        ("user","{text}")
    ]
)
```

### Set output parser

```python
parser = StrOutputParser()
```

#### Chaining

```python
chain = prompt_template | model | parser
```

```python
chain.invoke({"language": "urdu", "text":"I am hunger give me some pizza"})
```
